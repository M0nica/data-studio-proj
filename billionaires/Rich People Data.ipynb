```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import datetime
import re
```

```python
df = pd.read_excel('richpeople.xlsx')
df.head()
```

```python
recent_df = df[df['year'] == 2014]
```

```python
recent_df.head()
```

```python
recent_df.columns
```

```python
his = recent_df['networthusbillion'].hist(range=[0, 45])


his.set_title('Distribution of Wealth Amongst Billionaires')
his.set_xlabel('Wealth(Billions)')
his.set_ylabel('# of Billionnaires')
```

```python
his = recent_df['networthusbillion'].sort_values(ascending=False).head(20).hist(bins=20)


his.set_title('Distribution of Wealth Amongst Top 20 Richest Billionaires')
his.set_xlabel('Wealth(Billions)')
his.set_ylabel('# of Billionnaires')
```

```python
fig, ax = plt.subplots()
# need to remove negative ages from this data! -- resolved itself by using newer data sert 
recent_df.plot(kind='scatter', x='age', y='networthusbillion', ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.set_title("Distribution of Wealth - Self-Made vs Inherited")
for category, group in recent_df.groupby('selfmade'):
    ax.plot(group['age'], group['networthusbillion'], marker='o', linestyle='', label=category, markeredgewidth=0, alpha=0.4)
ax.legend()
plt.grid(True)
ax.set_ylabel("Networth US Billions")
ax.set_xlabel("Age")
plt.savefig('SELFdistributionOfIntheritance.pdf')
plt.savefig('SELFdistributionOfIntheritance.png')
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.set_title("Distribution of Wealth - Male vs Female")
for category, group in recent_df.groupby('gender'):
    ax.plot( group['age'], group['networthusbillion'], marker='o', linestyle='', label=category, markeredgewidth=0, alpha=0.4)
ax.legend()
plt.grid(True)
ax.set_ylabel("Networth US Billions")
ax.set_xlabel("Age")
plt.savefig('GENDERdistributionOfIntheritance.pdf')
plt.savefig('GENDERdistributionOfIntheritance.png')
```

```python
men_bill = recent_df[recent_df['gender'] == 'male']
```

```python
men_bill.head()
```

```python
women_bill = recent_df[recent_df['gender'] == 'female']
women_bill.head()
```

```python
colors = ['b', 'g', 'r', 'w']

fig, ax = plt.subplots(figsize=(10,10))
ax.set_title("Distribution of Wealth - Self-Made vs Inherited")
for category, group in women_bill.groupby('selfmade'):
    ax.plot(group['age'], group['networthusbillion'], marker='o', linestyle='', label=category + " (women)", markeredgewidth=0, alpha=0.25)
for category, group in men_bill.groupby('selfmade'):
    ax.plot(group['age'], group['networthusbillion'], marker='o', linestyle='', label=category + " (men)", markeredgewidth=0, alpha=0.25)
ax.legend()
plt.grid(True)
ax.set_ylabel("Networth US Billions")
ax.set_xlabel("Age")
plt.savefig('GENDERdistributionOfIntheritanceVSELF.pdf')
plt.savefig('GENDERdistributionOfIntheritanceVSELF.png')
```

```python
colors = ['b', 'g', 'r', 'w']

fig, ax = plt.subplots(figsize=(10,10))
ax.set_title("Distribution of Wealth - Self-Made vs Inherited")
for category, group in women_bill.groupby('selfmade'):
    ax.plot(group['age'], group['networthusbillion'], marker='o', linestyle='', label=category + " (women)", markeredgewidth=0, alpha=0.25)
ax.set_ylim([0, 80])
ax.legend()
plt.grid(True)
ax.set_ylabel("Networth US Billions")
ax.set_xlabel("Age")
```

Finding - Women billionaires are more likely to inherit their wealth whereas men are more likely to be self-made billonaires. There are a lot of successful self-made male billionnaires (range from 1-80 billion dollars) whereas self made women billionnaires all fall in the range of 1 to 10 billion dollars. 

Also, virtually no women billionaires that have over 40 billion dollars - whereas there are several male outliers that do.
