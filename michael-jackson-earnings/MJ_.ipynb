```python

# workon dataanalysis - my virtual environment
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('MichaelJacksonEarnings.csv')
```

```python
df.head()
```

```python
df.plot(x='Year', y='Earnings for Chart')
```

```python
ax = df.plot(kind= "bar", x='Year', y='Earnings for Chart', figsize=(20,10))
ax.legend_.remove()
ax.set_ylabel("Earnings in Millions")
ax.set_title("Michael Jackson's Earnings Over Time")
plt.savefig('MJEarningsOverTime.pdf')
```

```python
ax = df.plot(kind= "barh", x='Year', y='Earnings for Chart', figsize=(20,10))
ax.legend_.remove()
ax.set_ylabel("Earnings in Millions")
ax.set_title("Michael Jackson's Earnings Over Time")
plt.savefig('MJEarningsOverTime.pdf')
```
