```python
import datetime
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
social_df = pd.read_csv("NYT_Social_hitsByQuarter", encoding='latin-1')
```

```python
social_df.head()
```

```python
social_df['full_startdate'] = pd.to_datetime(social_df['full_startdate'])
```

```python
social_df['year'] = pd.to_datetime(social_df['year'])
```

```python
fig, ax = plt.subplots(figsize=(20,10))
for sitename, selection in social_df.groupby('social_site'):
    ax.plot(selection['year'], selection['hits'],label=sitename)
    
ax.legend(loc=2)
```
